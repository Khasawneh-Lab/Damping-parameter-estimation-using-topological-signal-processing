```python

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
from scipy.integrate import odeint
from PyEMD import EMD, EEMD
```

```python

def TDOF_SMD(time, IC, parameters):
    import numpy as np
    
    mu_c1, mu_v1, mu_q1 = parameters[0], parameters[1], parameters[2]
    k1, m1 = parameters[3], parameters[4]
    mu_c2, mu_v2, mu_q2 = parameters[5], parameters[6], parameters[7]
    k1, m1 = parameters[8], parameters[9]
    g = parameters[10]
    
    def f(state, t):
        x1, x1d, x2, x2d = state  # unpack the state vector
        DF1 = (mu_v1 + mu_v2)*x1d - mu_v2*x2d
        SF1 = (k1 + k2)*x1 - k2*x2
        DF2 = mu_v2*(x2d - x1d)
        SF2 = k2*(x2 - x1)
        state_derivates = [x1d,
                           -(1/m1)*(DF1 + SF1),
                           x2d,
                           -(1/m2)*(DF2 + SF2)]
                           
        return state_derivates[0], state_derivates[1], state_derivates[2], state_derivates[3]  # derivatives
    
    states = odeint(f, IC, time)
    return states

```

```python

fs = 200 #sampling frequency (Hz)
t = np.arange(0,4,1/fs)
mu_c, mu_v, mu_q = 0.0, 0.4, 0.0
dp1 = [mu_c, mu_v, mu_q]
k1, m1 = 20, 1

mu_c, mu_v, mu_q = 0.0, 0.5, 0.0
dp2 = [mu_c, mu_v, mu_q]
k2, m2 = 20, 1

g = 9.81
params = [dp1[0], dp1[1], dp1[2], k1, m1, 
          dp2[0], dp2[1], dp2[2], k2, m2,  
          g]

Param_SNR = 1000
SNR = 1000

Initial_Conditions = [0, 0, 0, 0]
states = TDOF_SMD(t, Initial_Conditions, params)
x1, x2 = states[:,0], states[:,1]




```
